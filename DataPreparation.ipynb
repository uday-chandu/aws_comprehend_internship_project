```python
import pandas as pd
import os,glob,re,nltk
from sklearn.utils import shuffle
```

```python
path=os.listdir('bbc')
text=[]
topic=[]
for folder in path:
    files=glob.glob('bbc/'+folder+'/*.txt')
    for file in files[:300]:
        topic.append(folder)
        data=open(file,'r').read()
        text.append(data)
print(text[:50])
print(topic[:100])
```

```python
text=pd.DataFrame(text,columns=['text'])
topic=pd.DataFrame(topic,columns=['topic'])
```

```python
df=pd.concat([topic,text],axis=1)
df = shuffle(df)
```

```python
#removing the '\n' from the text
df['text']=df['text'].str.replace('\n',' ')
```

```python
df.to_csv('training_set.csv',header=True,index=False)
```
